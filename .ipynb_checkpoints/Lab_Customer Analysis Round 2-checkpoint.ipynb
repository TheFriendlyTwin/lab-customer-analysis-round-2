```python
# Funtion that cleans NaN values, replacing it for 0 when it's a numerical column and dropping it when it's a categorical column
def clear_NaN(df):
    # Fill NaN values with 0 in numerical columns
    numerical_columns = df.select_dtypes(include='number').columns
    df[numerical_columns] = df[numerical_columns].fillna(0)
    
    # Drop categorical columns with NaN values
    categorical_columns = df.select_dtypes(include='object').columns
    df_dropped = df.dropna(subset=categorical_columns)

    return df_dropped
```

```python
#Checking with a copy before replacing in the original dataframe, if the NaN were cleared
df_clean = clear_NaN(df)
df_clean
```

```python
#Couting the NaN per column
df_clean.isna().sum()
```

```python
#Replacing the original dataframe with a cleared NaN dataframe

df = df_clean
df
```

```python
#For the numerical variables, check the multicollinearity between the features. Please note that we will use the column total_claim_amount later as the target variable.

# Check for correlations on the numerical data:

correlations_matrix = numerical.corr() # method="spearman"
correlations_matrix
```

```python
# From this information, we can check for the features with the highest correlations with the target (total_claim_amount):

correlations_matrix["Total Claim Amount"].sort_values(ascending=False)
```

```python
# You can write filters to help you understand these correlations:

correlations_matrix[(correlations_matrix['Total Claim Amount'] > .2) | (correlations_matrix['Total Claim Amount'] < -.2)]['Total Claim Amount']
```

```python
# Checking for multicolinearity between variables:
# rule of thumb: if two features have a correlation > .9, you might want to drop the one with smaller correlation with the target

mask = np.zeros_like(correlations_matrix)
mask[np.triu_indices_from(mask)] = True # optional, to hide repeat half of the matrix

fig, ax = plt.subplots(figsize=(10, 8))
ax = sns.heatmap(correlations_matrix, mask=mask, annot=True)
plt.show()
```

```python
#Drop one of the two features that show a high correlation between them (greater than 0.9). Write code for both the correlation matrix and for seaborn heatmap. If there is no pair of features that have a high correlation, then do not drop any features

# Create a heatmap using Seaborn
# Set the figure size and aspect ratio
fig, ax = plt.subplots(figsize=(10, 8))
sns.heatmap(correlations_matrix, annot=True, cmap='coolwarm', ax=ax)

# Add labels and title to the plot
plt.title("Correlation Heatmap")
plt.show()

#Because there is no pair of features that have a high correlation, we won't be dropping any feature.
```
